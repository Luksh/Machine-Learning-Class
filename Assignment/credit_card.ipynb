```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Amount"])
plt.title("Amount distribution with outliners")
plt.show()
```

```python
df["TimeAmount"]=df["Time"]*df["Amount"]
df.head()
```

```python
amount_bins=[0,100,200,300,400,500]
amount_labels=["Very Small","Small","Medium","Large","Very Large"]
df["AmountGroup"]=pd.cut(df["Amount"],bins=amount_bins,labels=amount_labels)
df.head()
```

```python
columns_to_keep=["Time","Amount","Class"]
credit_card_data=df[columns_to_keep]
credit_card_data.head()
```

```python
time_counts=credit_card_data["Time"].value_counts()

plt.figure(figsize=(8,6))
plt.pie(time_counts,labels=time_counts.index,autopct="%1.1f%%")
plt.axis("equal")
plt.title("Proportion of credit cards by time")
plt.show()
```

```python
cor_matrix=credit_card_data.corr()

plt.figure(figsize=(8,6))
sns.heatmap(cor_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation matrix heatmap")
plt.show()
```
