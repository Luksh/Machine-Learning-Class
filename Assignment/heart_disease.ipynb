```python
df["cp+ca"]=df["cp"]+df["ca"]
df.head()
```

```python
age_bins=[0,16,32,48,64,80]
age_labels=["Child","Young Adult","Adult","Middle-Age","Senior"]
df["AgeGroup"]=pd.cut(df["age"],bins=age_bins,labels=age_labels)
df.head()
```

```python
chol_bins=[0,150,300,450,float("inf")]
chol_labels=["Low","Medium","High","Very High"]
df["CholCategory"]=pd.cut(df["chol"],bins=chol_bins,labels=chol_labels)
df.head()
```
