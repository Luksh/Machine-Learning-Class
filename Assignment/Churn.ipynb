```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Intl Calls"])
plt.title("International Call distribution with outliners")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df["VMail Plan"]=le.fit_transform(df["VMail Plan"])
df.head()
```

```python
df["ToTal Mins"]=df["Day Mins"]+df["Eve Mins"]+df["Night Mins"]
df.head()
```

```python
account_bins=[0,50,100,150]
account_labels=["Small","Medium","Large"]
df["Account Length"]=pd.cut(df["Account Length"],bins=account_bins,labels=account_labels)
df.head()
```

```python
vmail_bins=[0,15,30,45,float("inf")]
vmail_labels=["Low","Medium","High","Very High"]
df["VMail Category"]=pd.cut(df["VMail Message"],bins=vmail_bins,labels=vmail_labels)
df.head()
```

```python
columns_to_keep=["Day Calls","Eve Calls","Night Calls"]
churn_data=df[columns_to_keep]
churn_data.head()
```

```python
day_calls_counts=churn_data["Day Calls"].value_counts()

plt.figure(figsize=(8,6))
plt.pie(day_calls_counts,labels=day_calls_counts.index,autopct="%1.1f%%")
plt.axis("equal")
plt.title("Proportion of Day Calls")
plt.show()
```

```python
cor_matrix=churn_data.corr()

plt.figure(figsize=(8,6))
sns.heatmap(cor_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation matrix heatmap")
plt.show()
```
