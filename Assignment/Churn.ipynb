```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df["VMail Plan"]=le.fit_transform(df["VMail Plan"])
df.head()
```

```python
df["ToTal Mins"]=df["Day Mins"]+df["Eve Mins"]+df["Night Mins"]
df.head()
```

```python
account_bins=[0,50,100,150]
account_labels=["Small","Medium","Large"]
df["Account Length"]=pd.cut(df["Account Length"],bins=account_bins,labels=account_labels)
df.head()
```

```python
vmail_bins=[0,15,30,45,float("inf")]
vmail_labels=["Low","Medium","High","Very High"]
df["VMail Category"]=pd.cut(df["VMail Message"],bins=vmail_bins,labels=vmail_labels)
df.head()
```
