```python
age_bins=[0,16,32,48,64,80]
age_labels=["Child","Young Adult","Adult","Middle-Age","Senior"]
df["AgeGroup"]=pd.cut(df["Age"],bins=age_bins,labels=age_labels)
df.head()
```

```python
bmi_bins=[0,18.5,24.9,50,float("inf")]
bmi_labels=["Underweight","Normal","Overweight","Illness"]
df["FareCategory"]=pd.cut(df["BMI"],bins=bmi_bins,labels=bmi_labels)
df.head()
```
