```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["sex"]=le.fit_transform(df["sex"])
df.head()
```

```python
le=LabelEncoder()
df1["sex"]=le.fit_transform(df1["sex"])
df1.head()
```

```python
age_bins=[0,16,32,48,64,80]
age_labels=["Child","Young Adult","Adult","Middle-Age","Senior"]
df["AgeGroup"]=pd.cut(df["age"],bins=age_bins,labels=age_labels)
df.head()
```

```python
age_bins=[0,16,32,48,64,80]
age_labels=["Child","Young Adult","Adult","Middle-Age","Senior"]
df1["AgeGroup"]=pd.cut(df1["age"],bins=age_bins,labels=age_labels)
df1.head()
```

```python
columns_to_keep=["sex","age","health"]
student_data=df[columns_to_keep]
student_data.head()
```

```python
columns_to_keep=["sex","age","health"]
student_data1=df1[columns_to_keep]
student_data1.head()
```

```python
age_counts=student_data["age"].value_counts()
plt.figure(figsize=(8,6))
plt.pie(age_counts,labels=age_counts.index,autopct="%1.1f%%")
plt.axis("equal")
plt.title("Proportion of students by Age")
plt.show()
```

```python
age_counts1=student_data1["age"].value_counts()
plt.figure(figsize=(8,6))
plt.pie(age_counts1,labels=age_counts1.index,autopct="%1.1f%%")
plt.axis("equal")
plt.title("Proportion of students by Age")
plt.show()
```

```python
cor_matrix=student_data.corr()
plt.figure(figsize=(8,6))
sns.heatmap(cor_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation matrix heatmap")
plt.show()
```

```python
cor_matrix1=student_data1.corr()
plt.figure(figsize=(8,6))
sns.heatmap(cor_matrix1,annot=True,cmap="coolwarm")
plt.title("Correlation matrix heatmap")
plt.show()
```
