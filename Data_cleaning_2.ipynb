```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("titanic.csv")
```

```python
import seaborn as sns
#Check for outliners in "Age" Column
print("Age Statistics:")
print(df["Age"].describe())
```

```python
#Plot a box plot to visualize outliners in "Age"
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Age"])
plt.title("Age distribution with outliners")
plt.show()
```

```python
#Check for outliners in "Fare" Column
print("Fare Statistics")
print(df["Fare"].describe())
```

```python
#Plot a box plot to visualize outliners in "Fare"
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Fare"])
plt.title("Fare distribution with outliners")
plt.show()
```

```python
#Import required library
from sklearn.preprocessing import LabelEncoder

#Create label encoder object
le=LabelEncoder()

#Encode "Sex" Column
df["Sex"]=le.fit_transform(df["Sex"])

#Encode "Embarked" Column
df["Embarked"]=le.fit_transform(df["Embarked"])

df.head()
```

```python
#Create a new column "FamilySize" by adding "SibSp" and "Parch" Columns
df["FamilySize"]=df["SibSp"]+df["Parch"]+1
df.head()
```
